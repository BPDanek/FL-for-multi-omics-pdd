```python
import pandas as pd

internal_results_path = '/Users/benjamindanek/Code/federated_learning_multi_modality_ancestry/multi_modality_fl/results/generated_figures_tables/k_averaged_results_table_k_all PPMI Uniform Stratified test.csv'
external_results_path = '/Users/benjamindanek/Code/federated_learning_multi_modality_ancestry/multi_modality_fl/results/generated_figures_tables/k_averaged_results_table_k_all PDBP Uniform Stratified test.csv'
```

```python
internal_data = pd.read_csv(internal_results_path)
extermal_data = pd.read_csv(external_results_path)
```

```python
# dataset = "PPMI"
# baseline_selected = internal_data[internal_data['num_clients'] == 0]
# baseline_selected['Dataset'] = 'PPMI'

# internal_selected = internal_data[internal_data['num_clients'] == 2]
# internal_selected['Dataset'] = 'PPMI'

# combined = pd.concat([baseline_selected, internal_selected])

dataset = "PDBP"
baseline_selected = extermal_data[internal_data['num_clients'] == 0]
baseline_selected['Dataset'] = 'PDBP'

external_selected = extermal_data[internal_data['num_clients'] == 2]
external_selected['Dataset'] = 'PDBP'

combined = pd.concat([baseline_selected, external_selected])
```

```python
combined['algorithm_name'] = combined['algorithm_name'].apply(lambda x: x.replace('Classifier', '') if type(x) == str else x)
combined_clean = combined.set_index(['Dataset', 'algorithm_name']).drop(columns=['num_clients', 'split_method', 'val_name', 'fold_idx', 'num_clients_std', 'split_method_std', 'val_name_std', 'fold_idx_std', 'algorithm_name_std'])
combined_clean = combined_clean.round(3)
display(combined_clean)
```

```python
vanilla = []
for c in combined_clean.columns:
    if '_std' not in c:
        vanilla.append(c)
    

for v in vanilla:
    c = v + '_std'
    combined_clean[v] = combined_clean[v].astype(str) + '±' + combined_clean[c].astype(str)
    combined_clean = combined_clean.drop(columns=[c])

display(combined_clean)
```
